```python
import pandas as pd
import matplotlib.pyplot as plt

# data 로드
file_path = "./sentiment_analysis.json"
data = pd.read_json(file_path)

# sentiment_score를 숫자로 변환
data["sentiment_score"] = pd.to_numeric(data["sentiment_score"], errors="coerce")

# 분기별 평균 sentiment_score 계산
#단순 평균
quarterly_sentiment = data.groupby("quarter")["sentiment_score"].mean().sort_index()

#가중평균
#quarterly_sentiment = data.groupby("quarter").apply(lambda x: (x["sentiment_score"] * x["sentiment_score"].count()).sum() / x["sentiment_score"].count()).sort_index()

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(quarterly_sentiment.index, quarterly_sentiment.values, marker="o", linestyle="-")
plt.xlabel("Quarter")
plt.ylabel("Average Sentiment Score")
plt.title("Quarterly Sentiment Score Distribution")
plt.xticks(rotation=45)
plt.xticks(ticks=range(0, len(quarterly_sentiment), 2), rotation=45)
plt.grid()
plt.show()
```
